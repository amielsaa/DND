## calculate mahalanobis:

```python
import pandas as pd 
```

```python
import numpy as np
```

```python
df = pd.read_csv("/content/df_openAI_test2.csv")
```

```python
headers = list(df.columns)
headers.pop(0)
```

```python
headers
```

```python
def mahalanobis(x=None, data=None, cov=None):

    x_mu = x - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(x_mu, inv_covmat)
    mahal = np.dot(left, x_mu.T)
    return mahal.diagonal()
```

```python
df['mahalanobis'] = mahalanobis(x=df[headers], data=df[headers])
```

```python
df
```

```python
from scipy.stats import chi2
#calculate p-value for each mahalanobis distance 
df['p'] = 1 - chi2.cdf(df['mahalanobis'], 117)
df
```

```python
df.to_csv("mahalanobis_openAI_test2.csv")
```

## calculate LOF:


```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import pairwise_distances
from sklearn.neighbors import LocalOutlierFactor
```

```python
X = df[headers]
```

```python
pairwise_distances(X,metric="manhattan") # to check the distances
```

```python
LOF = LocalOutlierFactor(n_neighbors = 2,metric="manhattan")
```

```python
LOF.fit_predict(X) #to predict the outliers
```

```python
arr = -LOF.negative_outlier_factor_ #let's multiply the negative scores by - and see the positive scores
```

```python
df["lof"] = arr
```

```python
df
```
